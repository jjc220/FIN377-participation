# New Q4

```python
from requests_html import HTMLSession
session=HTMLSession()
url = 'https://api.exchangeratesapi.io/history?start_at=2018-01-02&end_at=2019-12-31&symbols=JPY'

r=requests.get('https://api.exchangeratesapi.io/history?start_at=2018-01-02&end_at=2019-12-31&symbols=JPY')
r=requests.get(url)
df=pd.DataFrame(r.json())
df['EURJPY']=df['rates'].apply(lambda x: x.get('JPY'))
df
```
