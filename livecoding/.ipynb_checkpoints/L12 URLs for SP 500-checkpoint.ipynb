```python
import pandas as pd
from requests_html import HTMLSession
url='https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
session=HTMLSession()
r=session.get(url)
r
# we want the links!
```

```python
#blank force
r.html.links     #all ttthe relative links
r.html.absolute_links  #all of them can be "click on these"

#Get table, either works for getting constitents
r.html.find('table')
#r.html.find('table') [0]
table= r.html.find('#constituents') [0]
table.text #print the text
table.html  #printt the html


#blunt force the table!
table.absolute_links  #just that tables links...still too many, and can't "dumb filter"
```

```python
#let's loop over the rows
table.find('tr')[0].text  #header ...we don't want that!
table.find('tr')[1].text

table_rows=table.find('tr')

colinks=[]
for row in table_rows[1:]:
    a_link=list(row.find('td')[1].absolute_links)[0]
    colinks.append(a_link)
len(colinks)
```

```python
import pandas as pd
from requests_html import HTMLSession

#open page
url='https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
session=HTMLSession()
r=session.get(url)


#grab the URLAs
table= r.html.find('#constituents') [0]
table_rows=table.find('tr')

colinks=[]
for row in table_rows[1:]:
    a_link=list(row.find('td')[1].absolute_links)[0]
    colinks.append(a_link)
```

```python
import requests
def search_itunes(search_term):
    '''Simplified iTunes search'''
    
    base_url = 'https://itunes.apple.com/search'
    search_parameters = {'term': search_term}
    
    r = requests.get(base_url, params = search_parameters)
    
    results_df = pd.DataFrame(r.json()['results'])
    
    return results_df

search_itunes('billie eilish')      # one search at a time
search_itunes('father john misty') # "another one" - dj khaled

artists = ['billie eilish','father john misty'] # you can loop over them!

# download the results and save locally
for artist in artists:
    df = search_itunes(artist)
    print(len(df)) 
    
    
        
   
```
